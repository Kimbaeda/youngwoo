```python
# 백준 1927 최소 힙

import sys
import heapq

input = sys.stdin.readline

N = int(input())
heap = []

# Max Heap
for _ in range(N):
    num = int(input())
    if num != 0:
        heapq.heappush(heap, num)
    else:
        try:
            print(heapq.heappop(heap))
        except:
            print(0)
```

```python
# 백준 11279 최대 힙

import sys
import heapq

# input = sys.stdin.readline

N = int(input())
heap = []

#Min Heap
for _ in range(N):
    num = int(input())
    if num != 0:
        heapq.heappush(heap, (-num))
    else:
        try:
            print(-1 * heapq.heappop(heap))
        except:
            print(0)
```
