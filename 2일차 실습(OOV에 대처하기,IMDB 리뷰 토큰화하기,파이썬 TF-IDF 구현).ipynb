```python
import re, collections
```

```python
# BPE을 몇 회 수행할 것인지를 정함. 10회로 정함.
num_merges =10
```

```python
'''
BPE의 입력으로 사용하는 실제 단어 집합
</w>는 단어의 맨 끝에 붙이는 특수 문자이며, 각 단어는 글자(character) 단위로 분리
'''

dictionary = {'l o w ,</w>' : 5,
              'l o w e r </w>' : 2,
              'n e w e s t </w>' : 6,
              'w i d e s t </w>' : 3
              }
```

- unigram

- bigram

- trigram

우리들은 밥을 먹었고,

나는 공부중이다. /

```python
def get_stats(dictionary):
  # 유니그램의 pair들의 빈도수를 카운트
  pairs = collections.defaultdict(int)
  for word, freq in dictionary.items():
    symbols = word.split()
    for i in range(len(symbols)-1):
      pairs[symbols[i], symbols[i+1]] += freq
  print('현재 pair들의 빈도수 : ', dict(pairs))
  return pairs
```

```python
# 가장 빈도수가 높은 유니그램의 쌍을 하나의 유니그램으로 통합
def merge_dictionary(pair, v_in):
  v_out = {}
  bigram = re.escape(' '.join(pair))
  p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
  for word in v_in:
    w_out = p.sub(''.join(pair), word)
    v_out[w_out] = v_in[word]

  return v_out
```

```python
bpe_codes = {}
bpe_codes_reverse = {}
```

```python
# 가장 빈도수가 높은 유니그램의 쌍을 하나의 유니그램으로 통합하는 과정을 총 10 (num_merges=10) 회 반복
for i in range(num_merges):
  print(">> Step {0}".format(i+1))
  pairs = get_stats(dictionary)
  best = max(pairs, key=pairs.get)
  dictionary = merge_dictionary(best, dictionary)

  bpe_codes[best] = i
  bpe_codes_reverse[best[0] + best[1]] = best

  print(f'new merge: {best}')
  print(f'dictionary: {dictionary}')
```

## OOV에 대처하기

```python
# 글자 단위로 분리
def get_pairs(word):
    """Return set of symbol pairs in a word.
    Word is represented as a tuple of symbols (symbols being variable-length strings).
    """
    pairs = set()
    prev_char = word[0]
    for char in word[1:]:
        pairs.add((prev_char, char))
        prev_char = char
    return pairs
```

```python
# dictionary 10번 훈련 된 bpe_codes BPE 알고리즘으로  orig에 입력한 단어를 분리
def encode(orig):
    """Encode word based on list of BPE merge operations, which are applied consecutively"""

    word = tuple(orig) + ('</w>',)
    #display(Markdown("__word split into characters:__ <tt>{}</tt>".format(word)))
    print("__word split into characters:__ <tt>{}</tt>".format(word))

    pairs = get_pairs(word)    

    if not pairs:
        return orig

    iteration = 0
    while True:
        iteration += 1
        #display(Markdown("__Iteration {}:__".format(iteration)))
        print("__Iteration {}:__".format(iteration))

        print("bigrams in the word: {}".format(pairs))
        bigram = min(pairs, key = lambda pair: bpe_codes.get(pair, float('inf')))
        print("candidate for merging: {}".format(bigram))
        if bigram not in bpe_codes:
            #display(Markdown("__Candidate not in BPE merges, algorithm stops.__"))
            print("__Candidate not in BPE merges, algorithm stops.__")
            break
        first, second = bigram
        new_word = []
        i = 0
        while i < len(word):
            try:
                j = word.index(first, i)
                new_word.extend(word[i:j])
                i = j
            except:
                new_word.extend(word[i:])
                break

            if word[i] == first and i < len(word)-1 and word[i+1] == second:
                new_word.append(first+second)
                i += 2
            else:
                new_word.append(word[i])
                i += 1
        new_word = tuple(new_word)
        word = new_word
        print("word after merging: {}".format(word))
        if len(word) == 1:
            break
        else:
            pairs = get_pairs(word)

    # 특별 토큰인 </w>는 출력하지 않는다.
    if word[-1] == '</w>':
        word = word[:-1]
    elif word[-1].endswith('</w>'):
        word = word[:-1] + (word[-1].replace('</w>',''),)

    return word
```

## 파이썬 TF-IDF 구현

```python
import pandas as pd
from math import log
```

```python
docs = [
        '먹고 싶은 사과',
        '먹고 싶은 바나나',
        '길고 노란 바나나 바나나',
        '저는 과일이 좋아요'
]
```

```python
# docs 에서 doc 에서 w 띄어쓰기로 나눔.그 후 set로 중복제거 후 list에 저장.
# sort()로 정렬.
vocab = list (set(w for doc in docs for w in doc.split()))
vocab.sort()
```

```python
print(vocab)
```

```python
# 총 문서의 수
N = len(docs)
print(N)
```

 - tf(d,t) : 특정 문서 d에서의 특정 단어 t의 등장 횟수.

 - df(t) : 특정 단어 t가 등장한 문서의 수.

 - idf(d, t) : df(t)에 반비례하는 수.

 tf-idf = tf*idf

정리 잘된 사이트 링크 :

 https://thinkwarelab.wordpress.com/2016/11/14/ir-tf-idf-%EC%97%90-%EB%8C%80%ED%95%B4-%EC%95%8C%EC%95%84%EB%B4%85%EC%8B%9C%EB%8B%A4/
 

```python
# TF, IDF, 그리고 TF-IDF 값을 구하는 함수를 구현

def tf(t, d):
    return d.count(t)

def idf(t):
    df = 0
    for doc in docs:
        df += t in doc
    return log(N/(df + 1))

def tfidf(t, d):
    return tf(t,d)* idf(t)
```

```python
result = []
# tf 값 result에 저장.
for i in range(N): # 각 문서에 대해서 아래 명령을 수행
    result.append([])
    d = docs[i]
    for j in range(len(vocab)):
        t = vocab[j]        
        result[-1].append(tf(t, d))
# 데이터프레임으로 만들고 출력
tf_ = pd.DataFrame(result, columns = vocab)
tf_
```

```python
result = []
# idf 값 result에 저장
for j in range(len(vocab)):
    t = vocab[j]
    result.append(idf(t))
# 데이터프레임으로 만들고 출력
idf_ = pd.DataFrame(result, index = vocab, columns = ["IDF"])
idf_
```

```python
result = []
# tfidf값 result에 저장
for i in range(N):
  result.append([])
  d = docs[i]
  for j in range(len(vocab)):
    t = vocab[j]
    result[-1].append(tfidf(t,d))
# 데이터프레임으로 만들고 출력
tfidf_ = pd.DataFrame(result, columns= vocab)
tfidf_
```

문장의 시작에 (sos), 문장의 끝에 (eos) 를 넣기도 합니다. 각각 Start-of-sentence, End-of-sentence 의 약자입니다. 보통 sos는 안 넣지만 eos는 많이 넣습니다. 문장이 끝났다는 것 까지 학습하도록 하기 위해서는 eos를 넣어 주는 것이 좋습니다. 위의 UNK, EOS, SOS 등은 특수한 토큰입니다.

```python
# 0 <pad>
# 1 <sos>
# 2 <unk>
# index 0,1,2 에 <pad>, <sos>, <unk>
for index, token in enumerate(('<pad>','<sos>','<unk>')):
  index_to_word[index] = token
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
#- 단어의 수를 카운트하는 사이킷런의 카운트벡터라이저입니다.
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(x_train)
# fit_transform : 학습할 때와 동일한 기반 설정으로 동일하게 테스트 데이터를 변환해야 하는 것

#- 카운트벡터라이저의 결과로부터 TF-IDF 결과를 얻습니다.
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

#- 나이브 베이즈 분류기를 수행합니다.
#- X_train은 TF-IDF 벡터, y_train은 레이블입니다.
clf = MultinomialNB().fit(X_train_tfidf, y_train)
```

```python
def tfidf_vectorizer(data):
  data_counts = count_vect.transform(data)
  data_tfidf = tfidf_transformer.transform(data_counts)
  return data_tfidf
```

```python
y_pred = clf.predict(tfidf_vectorizer(x_test))
print(metrics.classification_report(y_test, y_pred))
```
