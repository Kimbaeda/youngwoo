```python
import sys
import math
# input = sys.stdin.readline

sum_g = 0

def sum_gcd(arr):
    global sum_g
    for i in range(len(arr)-1):
        for j in range(i+1,len(arr)):
            sum_g += math.gcd(arr[i], arr[j])
    return sum_g


n = int(input())

for _ in range(n):
    sum_g = 0
    N, *arr = map(int, input().split())
    print(sum_gcd(arr))
```
